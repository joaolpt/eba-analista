## while

O loop `while` em python é uma estrutura de controle que permite executar um bloco de código repetidamente enquanto uma condição específica for verdadeira. A sintaxe básica é:

```python
while condição:
    # bloco de código
```
Características:
- Condição: O loop continua enquanto a condição for verdadeira. Se a condição for falsa desde o início, o bloco de código não será executado.
- Atualização da Condição: É importante garantir que a condição eventualmente se torne falsa, para evitar loops infinitos.
- Uso: É útil em situações onde o número de iterações não é conhecido previamente, como ao esperar por uma entrada do usuário ou ao monitorar o status de um processo.

````python
numero = 5

while numero > 0:
    print(numero)
    numero = numero - 1
````

As vezes quando escrevemos um código fazemos operação do tipo:
1. `a = a + 1`
2. `b = b - 1`

`python` prove algumas estruturas para facilicar isso:

1. `a = a + b` é equivalente a `a += b`. Note que isso significa que faremos a soma de `a` com `b` e salvaremos o resultado em `a`,
2. `a = a - b` é equivalente a `a -= b`. Note que isso significa que faremos a diferença de `a` com `b` e salvaremos o resultado em `a`.
3. `a = a * b` é equivalente a `a *= b`. Note que isso significa que faremos a múltiplicação de `a` com `b` e salvaremos o resultado em `a`.
3. `a = a / b` é equivalente a `a /= b`. Note que isso significa que faremos a divisão de `a` com `b` e salvaremos o resultado em `a`.

### while/else

Em Python, um loop while pode ser seguido por um bloco else, que é executado quando o loop termina normalmente, ou seja, quando a condição do while se torna falsa. O bloco else não é executado se o loop for interrompido por uma instrução break

## List comprehension

List comprehension é uma forma concisa e elegante de criar listas em Python. Ao invés de usar loops tradicionais, a list comprehension permite que você crie uma nova lista aplicando uma expressão a cada item de uma lista existente, de forma compacta. Exemplo báscio:

```py
[nova_expressao for item in lista]
```

### Filtros com list comprehension:
Você também pode adicionar condições (filtros) para incluir apenas os itens que atendem a um critério.

### Dicionários

Assim como em listas, dict comprehension permite criar dicionários de maneira concisa e legível. Ele segue a mesma ideia de transformar ou filtrar dados, mas, em vez de gerar uma lista, gera um dicionário.
