# Operações com Arrays

```python

a = np.array([5, 9, 1])
b = np.array([1, 7, 8])
soma = a + b
sub = a - b
mult = a * b
divi = a / b
pot = a ** b
print(soma)
print("-" * 30)
print(sub)
print("-" * 30)
print(mult)
print("-" * 30)
print(divi)
print("-" * 30)
print(pot)
```

```python
seno = np.sin(a)
raiz = np.sqrt(a)
print(seno)
print("-" * 60)
print(raiz)
```

# Operação com Matrizes

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[10, 9], [8, 7]])
print(m1)
print("-" * 60)
print(m2)
```

```python
soma = m1 + m2
sub = m1 - m2
mult = m1 * m2
divi = m1 / m2
pot = m1 ** m2
print(soma)
print("-" * 60)
print(sub)
print("-" * 60)
print(mult)
print("-" * 60)
print(divi)
print("-" * 60)
print(pot)
```

# Operações Lógicas

```python
estoque = np.array([10, 20, 5]) #Quantidade em estoque
disponibilidade = np.array([1, 10, 2]) #Dias que demora para entregar 

print(np.where(estoque > 5, "OK", "NOT OK")) #Supondo que o min de estoque é 6
print("-" * 60)
print(np.where(disponibilidade <= 3, "OK", "NOT OK")) #e que a disponibilidade é menor q 3
print("-" * 60)
print(np.where((estoque > 5) & (disponibilidade <= 3), "OK", "NOT OK"))#Unindo as duas logicas
print("-" * 60)
print(np.where((estoque > 5) | (disponibilidade <= 3), "OK", "NOT OK"))#Unindo as duas logicas, porem apenas 1 deve ser verdadeiro ou no caso 'OK'
```

```python
#Exemplo
notas = np.array([45, 75, 60, 90, 50, 80, 30, 65])
print(np.where(notas >= 50, "Aprovada", "Reprovada"))
```

```python
# maior que 90 = A
# maior que 80 = B
# maior ou igual a 50 = C
# menor que 50 = Reprovado
np.select(
    [
        notas > 90,
        notas > 80,
        notas >= 50,
        notas < 50
    ],
    [
        "A",
        "B",
        "C",
        "Reprovado"
    ], default = "Erro"
)
```

# Álgebra Linear

```python
#2x + y = 8
#3x + 4y = 18
A = np.array([[2, 1], [3, 4]])
B = np.array([8, 18])

np.linalg.solve(A, B)
```

```python
#Determinnante
np.linalg.det(A)
```

```python
#Matriz Inversa
np.linalg.inv(A)
```

```python
#50x + 20y = 4000
#30x + 40y = 3600
C = np.array([[50, 20], [30, 40]])
D = np.array([4000, 3600])

np.linalg.solve(C, D)
```
