```python
##Agrupamento
df_mean = df.groupby("Idade")["Salário"].mean()
display(df_mean)
```

```python
df_sum = df.groupby("Idade")["Salário"].sum()
display(df_sum)
```

# Combinado com Numpy

```python
df['Idade'].describe()
```

```python
import numpy as np
df["Quartis"] = np.select(
    [
        df['Idade'] <= 26,
        df['Idade'] <= 41.5,
        df['Idade'] <= 51
    ],
    [
        "Q1",
        "Q2",
        "Q3"
    ], default="Q4"
    )
display(df)
```
