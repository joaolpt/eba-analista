# Gráficos com Matplotlib, seaborn e plotly

```python
import matplotlib.pyplot as plt

plt.plot(
    [1, 2, 3],
    [5, 3, 1],
    marker = "o"
)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Exemplo Básico")
```

```python
import pandas as pd
df = pd.read_csv("dados.csv")
display(df.head())
```

```python
plt.scatter(
    df["Idade"],
    df["Salário"],
)
plt.xlabel("Idade")
plt.ylabel("Salário")
```

```python
df_agrupado = df.groupby("Profissão")["Salário"].mean()

df_agrupado.index
display(df_agrupado)

plt.bar(
    df_agrupado.index,
    df_agrupado
)
```

```python
plt.hist(df["Salário"], bins=10);
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
```

```python
display(tips)
```

```python
sns.scatterplot(tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.violinplot(tips, y="tip", x="sex")
```

```python
sns.violinplot(tips, y="tip", x="sex", hue="smoker")
```

```python
display(flights)
```

```python
flights_pivot = flights.pivot(index="year", columns="month", values="passengers")
display(flights_pivot)

sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)
```

```python
import plotly.express as px

px.scatter(tips, x="total_bill", y= "tip", color="smoker")
```
